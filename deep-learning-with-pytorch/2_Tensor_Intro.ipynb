```python
#Tensor addition and multiplication
#Various ways you can perform tensor addition
a = torch.rand(2,2) 
print("a:",a)
b = torch.rand(2,2)
print("b:",b)
c = a + b
print("c:",c)
```

```python
d = torch.add(a,b)
print("d:",d)
```

```python
#For in-place addition
a.add_(5)
print("a:",a)
```

```python
#Multiplication of different tensors
test1 = a*b
print("test1:",test1)
test2 = a.mul(b)
print("test2:",test2)
test3 = a.mul_(b)
print("test3:",test3)
```
